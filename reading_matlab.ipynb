```python
"""
Created on Tues Nov 5 2024

@author: Lucas Berens
"""

from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt

#   Hyperslicer, install with:
#   >pip install mpl_interactions["jupyter"]
from mpl_interactions import hyperslicer

%matplotlib ipympl
```

```python
# Mouse 2
contents = loadmat('data/03-07-2024_Mouse_1.mat')
images = contents.get('results')

EPR    = images[0][0][0] # last index: 0=pO2, 1=PET TMR, 2=Corrected PET, 3=Ktrans, 4=Ve
TMR    = images[0][0][1]
cTMR   = images[0][0][2]
Ktrans = images[0][0][3]
Ve     = images[0][0][4]

EPR    = np.swapaxes(EPR,0,2)
TMR    = np.swapaxes(TMR,0,2)
cTMR   = np.swapaxes(cTMR,0,2)
Ktrans = np.swapaxes(Ktrans,0,2)
Ve     = np.swapaxes(Ve,0,2)

fig, (ax_EPR, ax_TMR, ax_cTMR, ax_Ktrans, ax_Ve) = plt.subplots(nrows=1, ncols=5, figsize=(15,5))
hyperslicer(EPR, vmin=0, vmax=50, ax=ax_EPR, play_buttons=True, play_button_pos="left", title='EPR pO2')
hyperslicer(TMR, vmin=0, vmax=3, ax=ax_TMR, play_buttons=True, play_button_pos="left", title='Uncorrected TMR PET')
hyperslicer(cTMR, vmin=0, vmax=3, ax=ax_cTMR, play_buttons=True, play_button_pos="left", title='Corrected TMR PET')
hyperslicer(Ktrans, vmin=0, vmax=0.2, ax=ax_Ktrans, play_buttons=True, play_button_pos="left", title='Ktrans DCE')
hyperslicer(Ve, vmin=0, vmax=1.0, ax=ax_Ve, play_buttons=True, play_button_pos="left", title='Ve DCE')
```
